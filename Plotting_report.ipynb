```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
x = np.linspace(-3, 3, 500)  # Input range
relu = np.maximum(0, x)  # Rectified Linear Unit (ReLU)
tanh = np.tanh(x)  # Hyperbolic Tangent (Tanh)
gelu = 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))  # Gaussian Error Linear Unit (GeLU)
elu = np.where(x > 0, x, np.exp(x) - 1)  # Exponential Linear Unit (ELU)
swish = x / (1 + np.exp(-x))  # Swish activation function
softplus = np.log(1 + np.exp(x))  # Softplus activation function
sigmoid = 1 / (1 + np.exp(-x))  # Sigmoid activation function
softmax = np.exp(x) / np.sum(np.exp(x))

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(x, relu, label="ReLU (Rectified Linear Unit)", linewidth=2)
plt.plot(x, tanh, label="Tanh (Hyperbolic Tangent)", linewidth=2)
plt.plot(x, gelu, label="GeLU (Gaussian Error Linear Unit)", linewidth=2)
# plt.plot(x, elu, label="ELU (Exponential Linear Unit)", linewidth=2)
# plt.plot(x, swish, label="Swish", linewidth=2)
# plt.plot(x, softplus, label="Softplus", linewidth=2)
plt.plot(x, sigmoid, label="Sigmoid", linewidth=2)
# plt.plot(x, softmax, label="Softmax", linewidth=2)

# Add grid, labels, legend, and title
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Common Activation Functions in Neural Networks", fontsize=20)
plt.xlabel("Input", fontsize=16)
plt.ylabel("Output", fontsize=16)
plt.legend(fontsize=14)
plt.grid(alpha=0.3)

plt.tick_params(axis='both', which='major', labelsize=12)

# Save the plot
plt.show()  # Display the plot
```
